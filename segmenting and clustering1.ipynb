## Part 1: Scraping Data 

```python
#importing libries

%matplotlib inline
import matplotlib.cm as cm
import matplotlib.colors as colors

#importing data frame
import pandas as pd
import numpy as np

#import packages and requests
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
```

```python
#defining wikipedia site to scrape data
url='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=890001695'
html=urlopen(url)
html=BeautifulSoup(html)  
html
```

```python
print(html.prettify())
```

```python
html.title
```

```python
html.title.text
```

```python
#identify the table that contains date
tbl=html.find_all('table',class_="wikitable sortable")
tbl
```

```python
#identifiying all the headers
headers=tbl[0].find_all('th')
headers
```

```python
#extracting the text with .text and loop thorugh each items
column=[i.text.strip() for i in headers]
column
```

```python
#get the data of the entire tables
data=tbl[0].find_all('td')
data
```

```python
#getting total number of rows the table contain
total_cell,total_row=len(data),len(data)/3
print('The table cointains:{}cells,which mean{}rows'.format(total_cell,total_row))
```

```python
#looping through the data 
col_1=[]

for i in range(0,total_cell,3):
    
    k=data[i].text.strip()
    col_1.append(k)
```

```python
print(col_1,end='')
```

```python
#looping thorugh the data with incremental
col_2=[]

for i in range(1,total_cell,3):
    
    k=data[i].text.strip()
    
    col_2.append(k)
```

```python
col_2
```

```python
#looping through the data with another increments
col_3=[]

for i in range(2,total_cell,3):
    
    k=data[i].text.strip()
    
    col_3.append(k)
```

```python
col_3
```

```python
#defining dataframe
tbl_df={}

for a,b in zip(column,[col_1,col_2,col_3]):
    
    tbl_df[a]=b #put the header and column into dictionary
```

```python
print(tbl_df['Postcode'],end='')
```

```python
#create dataframe and show the dataframe
tbl_postcode=pd.DataFrame(tbl_df)
tbl_postcode.head()
```

```python
#excluding the not assigned data
tbl=tbl_postcode.query("Borough !='Not Assigned'")
tbl
```

```python
#creating temporary table and group ib by header data
tbl_temp=tbl.groupby(['Postcode'])['Neighbourhood'].count().to_frame().reset_index(drop=False).sort_values(by='Neighbourhood',ascending=False)
tbl_temp
```

```python
#send query to postcode with the sample 
tbl.query("Postcode =='M8Y'")
```

```python
#defining df multi and query within the data
multi=tbl_temp.query("Neighbourhood > 1")['Postcode'].values
multi
```

```python
#defining df single and query
single=tbl_temp.query("Neighbourhood ==1")['Postcode'].values
single
```

```python
#spliting table with neighbourhood and postcode
t1=tbl.loc[tbl.Postcode.isin(multi),:].sort_values(by='Postcode').reset_index(drop=True)
t1
```

```python
#slipting table with neighbourhood and postcode
t2=tbl.loc[tbl.Postcode.isin(single),:].sort_values(by='Postcode').reset_index(drop=True)
t2
```

```python
#initiate all three list
codes=[]
boroughs=[]
neighbours=[]

for code in multi:
    
    table=t1.loc[t1.Postcode==code,:]
    
    code=np.unique(table.Postcode)
    codes.append(code[0])
    
    borough=np.unique(table.Borough)
    boroughs.append(borough[0])
    
    neighbour=table.Neighbourhood.to_list()
    neighbour=','.join(neighbour)
    neighbours.append(neighbour)
                    
```

```python
neighbours
```

```python
#create a new dat frame
tbl_multi=pd.DataFrame({'Postcode':codes,'Borough':boroughs,'Neighbourhood':neighbours})
tbl_multi
```

```python
#create a new dataframe
new_df=pd.concat([tbl_multi,t2],axis=0).sort_values(by='Postcode').reset_index(drop=True)
print(new_df.head())
```

```python
#query the new dataframe with the not assigned data for neighbourhood
new_df.query("Neighbourhood=='Not Assigned'")
```
